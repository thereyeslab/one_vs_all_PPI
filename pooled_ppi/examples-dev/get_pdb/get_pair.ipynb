```python
# Needs latest: pip install pooled-ppi @ git+https://github.com/jurgjn/pooled-ppi@main
import pooled_ppi
from pooled_ppi.core import *
from pprint import pprint
```

```python
# Find interactions with @protein_id; filter for AF3 quality using @af3_pair_score
protein_id = "p43603"
yp = pooled_ppi.PooledPredictionsDb('/cluster/work/beltrao/jjaenes/25.12_pooled-ppi-yeast/data-26.03')
pairs_ = yp.pairs.query('(af3_id1 == @protein_id | af3_id2 == @protein_id) & (af3_pair_score > 995)')
pairs_
```

```python
# Write PDB for every pair
for i, r in pairs_.iterrows():
    ids = [f'{r["name"]}_{r.af3_id1}', f'{r["name"]}_{r.af3_id2}',]
    path = f'{r.af3_id1}_{r.af3_id2}.pdb'
    #print(ids, path)
    print('Writing', path)
    yp.save_ids(ids, path)
```
