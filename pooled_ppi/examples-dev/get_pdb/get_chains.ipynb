```python
# Needs latest: pip install pooled-ppi @ git+https://github.com/jurgjn/pooled-ppi@main
import pooled_ppi
from pooled_ppi.core import *
from pprint import pprint
```

```python
# Get pairs with the protein-of-interest (p0c2j7) from the table of all pairs
protein_id = 'P0C2J7'
pp = pooled_ppi.PooledPredictionsDb('/cluster/work/beltrao/jjaenes/25.12_pooled-ppi-yeast/data-26.03')
pairs_ = pp.pairs.query('uniprot_id1 == @protein_id | uniprot_id2 == @protein_id')
pairs_
```

```python
# Get unique pools that have the protein-of-interest
pools_ = pairs_['name'].drop_duplicates(keep='first').tolist()
pools_
```

```python
# Build foldcomp id-s
ids = [f'{pool_id}_{protein_id.lower()}' for pool_id in pools_]
printlen(ids, f'pools contain {protein_id}')
```

```python
# Write out first n chains separately; files have equal size as pdb is a fixed-width format..
for id in ids[:5]:
    pp.save_ids([id,], f'{id}.pdb')
!ls -lh *.pdb
```
